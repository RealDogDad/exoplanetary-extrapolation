```python
from sklearn.datasets import fetch_lfw_people
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
faces.keys()
```

```python
faces.data.shape    # 2914 = 62x47   data is the flattend images.
```

```python
faces.images.shape   #1348 pictures, each picture is represented by 2D array of 62x47 pixel values
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.imshow(faces.images[0], cmap="bone")
```

```python
fig, ax = plt.subplots(3,5)      
# objected oriented interface to display multiple pictures, 3 rows and 5 columns of pictures

for i, axi in enumerate(ax.flat):     # i loops from 0 to 14, axi loops from 0th axis to 14th axis
    axi.imshow(faces.images[i], cmap="bone")
    axi.set(xticks=[], yticks=[],
           xlabel=faces.target_names[faces.target[i]])

plt.tight_layout()

# faces.target[i]   the index of the name of the ith image
# face.target_names[]   will give you the name for that particular index.
    
```

3. This question is about recognizing handwritten digits.
(a) Create the following image:
- First, load the dataset
- Second, display the image using object oriented interface (fig, ax = plt.subplots())


```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.keys()
```

```python
digits.target_names      # 10 classes 
```

```python
digits.target[999]
```

```python
digits.data.shape      # 1797 images, data shows the flattened image
```

```python
digits.images.shape     # 1797 images, each image is a 2D array of 8x8 pixel values.
```

```python
fig, ax =plt.subplots(8,8)      # 8 rows of images, 8 columns of images
for i, axi in enumerate(ax.flat):     # i loops from 0 to 63, axi loops from 0th axis to 63rd axis
    axi.imshow(digits.images[i], cmap="binary")
    axi.set(xticks=[], yticks=[])
```

```python
digits.images.shape     # each image is represented by 8x8 2D array.
# each image has 64 pixel values, each image has 64 "features"
# we want to reduce each image into just 2 features
# feature reduction
```

```python
from sklearn.manifold import Isomap     # Isomap is for feature reduction (dimensionality reduction)
iso = Isomap(n_components = 2)     # instantiation 
projection = iso.fit_transform(digits.data)    # trasnformtion of digits.data
```

```python
projection.shape       # 1797 rows, 1797 images, each image is just reduced to 2 features.
```

```python
plt.scatter(x=projection[:,0], y = projection[:,1], c = digits.target, 
            cmap = plt.get_cmap("cubehelix",10))
plt.colorbar(label="digit value")
plt.clim=(-0.5,9.5)
```

(a)	Filter the DataFrame to only include the rows corresponding to medal winners from 2016. (In other words, drop the rows if 'Medal'==NaN

(c)	Filter the DataFrame one more time to only include the records for the top five sports based on the number of medals in 2016. 

```python
olympics_2016 = olympic_winners[olympic_winners["Year"]==2016]
```

```python
olympics_2016["Sport"].value_counts()
```

```python
top_five=["Athletics", "Swimming", "Rowing", "Football", "Hockey"]
```

```python
olympics_top_five_2016 = olympics_2016[(olympics_2016["Sport"].isin(top_five))]
```

```python
olympics_top_five_2016.head()
```

```python
hpi_df = pd.read_csv("hpi_data_countries.txt", sep='\t')
hpi_df.head()
```

```python
g = sns.catplot(x="Region", data=hpi_df, kind="count", aspect=2)
g.set_xticklabels(rotation = 90)
```

```python
hpi_df["Wellbeing (0-10)(binned)"] = pd.cut(hpi_df["Wellbeing (0-10)"], bins=np.linspace(2.5,8,12))
hpi_df["Life Expectancy (years)(binned)"] = pd.cut(hpi_df["Life Expectancy (years)"], bins=np.linspace(45,85,9))
```

```python
hpi_df
```

```python
h = hpi_df.groupby("Life Expectancy (years)(binned)")["Wellbeing (0-10)(binned)"].value_counts().unstack()
h.fillna(0, inplace = True)
h.sort_index(ascending = False, inplace = True)
h
```

```python
sns.heatmap(h, cmap="Greens")
```
